## 2. Analisis Sentimen dengan Transformers

```python
list_tweets = df['full_text'].values.tolist()
list_tweets
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa")
```

```python
import re

corpus = []
for i in range(0, len(df)):
    tweet = re.sub(r'^https://t.co/[a-zA-Z0-9]*\s',' ', list_tweets[i])
    tweet = re.sub(r'\s+https://t.co/[a-zA-Z0-9]*\s',' ',tweet)
    tweet = re.sub(r'\s+https://t.co/[a-zA-Z0-9]*$',' ',tweet)
    tweet = tweet.lower()
    tweet = re.sub(r"\W"," ",tweet)
    tweet = re.sub(r"\d"," ",tweet)
    tweet = re.sub(r"\s+[a-z]\s+"," ",tweet)
    tweet = re.sub(r"\s+[a-z]$"," ",tweet)
    tweet = re.sub(r"^[a-z]\s+"," ",tweet)
    tweet = re.sub(r"\s+"," ",tweet)
    corpus.append(tweet)

corpus
```

```python
predict = pipe(corpus)
predict
```

```python
labels = []
total_pos = 0
total_net = 0
total_neg = 0
for i in range(len(predict)):
  labels.append(predict[i]['label'])
  if predict[i]['label'] == 'Positive':
      total_pos += 1
  elif predict[i]['label'] == 'Neutral':
      total_net += 1
  else:
      total_neg += 1

result = pd.DataFrame(labels, columns=['result'])
result['text'] = list_tweets
result['clean'] = corpus
result
```

```python
result.to_csv('result.csv')
print("Total Positive : ", total_pos)
print("Total Neutral : ", total_net)
print("Total Negative : ", total_neg)
```

```python
import matplotlib.pyplot as plt
import numpy as np

objects = ['Positive', 'Neutral','Negative']
y_pos = np.arange(len(objects))

plt.bar(y_pos,[total_pos, total_net, total_neg],alpha=0.5)
plt.xticks(y_pos,objects)
plt.ylabel('Number')
plt.title('Positive, Neutral, and Negative')

plt.show()
```
